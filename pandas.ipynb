```python
# primeiro sempre lembrar de importar
import pandas as pd
```

SERIES é um array unidimensional, funciona como uma coluna, que é rotulado por labels (representado abaixo pelo index 0, 1, 2)

```python
a = [1, 2, 3]
myvar = pd.Series(a)
print(myvar)
```

como dito antes, label é o rotulo da linha do series, então podemos mudar esse label, usando o argumento opcional index

```python
myvar = pd.Series(a, index = ['x', 'y', 'z'])
myvar
```

```python
# podemos acessar usando index de array normal
myvar['y']
```

Chave/Valor values também podem ser Series ja com o label aplicado na chave

```python
b = {'um': 1, 'dois': 2, 'tres': 3}
myvar2 = pd.Series(b)
print(myvar2)
```

DATAFRAMES são tables multidimensionais, algo como um arquivo csv

```python
data = {
    'calories': [200, 370, 890],
    'values': [4.50, 6.70, 10.0]
}

myvar = pd.DataFrame(data, index=['banana grande', 'pacote amendoim', 'salgadinho'])
myvar
```
